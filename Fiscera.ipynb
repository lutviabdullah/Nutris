```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Fiscera.csv')

# Display basic info
print("Dataset Shape:", df.shape)
print("\nFirst few rows:")
df.head()
```

```python
#Basic Exploration
# Check dataset info
print("Dataset Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nData Types:")
print(df.dtypes)

print("\nBasic Statistics:")
df.describe(include='all')
```

```python
#Cleaning data

# Create a copy for cleaning
df_clean = df.copy()

# Check for duplicates
print("Duplicate rows:", df_clean.duplicated().sum())

# Remove duplicates if any
df_clean = df_clean.drop_duplicates()
print("Shape after removing duplicates:", df_clean.shape)
```

```python
# Cek informasi kolom
print("Informasi dataset:")
df.info()
```

```python
# Fokus pada kolom yang ingin dianalisis
target_column = "Ukuran Teks, Jenis Font, dan ikon mudah dibaca? "

print("Statistik deskriptif:")
print(df[target_column].describe())

print(f"\nNilai unik: {sorted(df[target_column].unique())}")
```

```python
# Hitung frekuensi setiap nilai
value_counts = df[target_column].value_counts().sort_index()
print("Distribusi jawaban:")
print(value_counts)

# Persentase
percentage = (value_counts / len(df) * 100).round(2)
print("\nPersentase:")
print(percentage)
```

```python
#Analisis 1 : Keterbacaan Teks

# Visualisasi 1: Bar Chart
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
bars = plt.bar(value_counts.index, value_counts.values, color='skyblue', edgecolor='black')
plt.title('Distribusi Jawaban: Keterbacaan Teks & Ikon', fontsize=14, fontweight='bold')
plt.xlabel('Skor (1-5)', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
plt.xticks([3, 4, 5])

# Tambahkan nilai di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom')

# Visualisasi 2: Pie Chart
plt.subplot(1, 2, 2)
colors = ['#ff9999', '#66b3ff', '#99ff99']
plt.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90)
plt.title('Persentase Skor Keterbacaan', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Skor rata-rata
mean_score = df[target_column].mean()
median_score = df[target_column].median()
mode_score = df[target_column].mode()[0]

print(f"ANALISIS HASIL - Keterbacaan Teks, Font, dan Ikon")
print("=" * 50)
print(f"Skor Rata-rata: {mean_score:.2f}")
print(f"Median: {median_score}")
print(f"Modus: {mode_score}")
print(f"Total Responden: {len(df)}")
print(f"Standar Deviasi: {df[target_column].std():.2f}")
```

```python
#Analisis 2 : Kenyamanan Warna Tema

# Fokus pada kolom warna tema
warna_column = "Warna tema (#75B1DC ,  #5F90B3, #3F5F76) terasa nyaman untuk dilihat."

print(" STATISTIK DESKRIPTIF - KENYAMANAN WARNA TEMA")
print("=" * 60)

# Statistik dasar
print(f"Total Responden: {len(df)}")
print(f"Skor Rata-rata: {df[warna_column].mean():.2f}")
print(f"Median: {df[warna_column].median()}")
print(f"Modus: {df[warna_column].mode()[0]}")
print(f"Standar Deviasi: {df[warna_column].std():.2f}")
print(f"Range: {df[warna_column].min()} - {df[warna_column].max()}")

# Distribusi nilai
value_counts = df[warna_column].value_counts().sort_index()
percentage = (value_counts / len(df) * 100).round(2)

print(f"\n DISTRIBUSI NILAI:")
for score, count in value_counts.items():
    print(f"Skor {score}: {count} responden ({percentage[score]}%)")
```

```python
# Visualisasi 1: Bar Chart dengan warna tema
plt.figure(figsize=(12, 5))

# Warna yang sesuai dengan tema
theme_colors = ['#75B1DC', '#5F90B3', '#3F5F76']

plt.subplot(1, 2, 1)
bars = plt.bar(value_counts.index, value_counts.values,
               color=theme_colors, edgecolor='black', alpha=0.8)
plt.title('Distribusi Skor Kenyamanan Warna Tema', fontsize=14, fontweight='bold')
plt.xlabel('Skor Kenyamanan (1-5)', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
plt.xticks([3, 4, 5])

# Tambahkan nilai di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# Visualisasi 2: Pie Chart
plt.subplot(1, 2, 2)
plt.pie(value_counts.values, labels=[f'Skor {k}' for k in value_counts.index],
        autopct='%1.1f%%', colors=theme_colors, startangle=90)
plt.title('Persentase Skor Kenyamanan Warna', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Analisis 3

# Visualisasi 1: Bar Chart
plt.figure(figsize=(12, 5))

# Warna untuk visualisasi
colors = ['#4CAF50', '#8BC34A', '#CDDC39', '#FFEB3B', '#FFC107']

plt.subplot(1, 2, 1)
bars = plt.bar(value_counts.index, value_counts.values,
               color=colors[2:5], edgecolor='black', alpha=0.8)
plt.title('Distribusi Skor Kemudahan Beranda', fontsize=14, fontweight='bold')
plt.xlabel('Skor Kemudahan (1-5)', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
plt.xticks([3, 4, 5])

# Tambahkan nilai di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# Visualisasi 2: Pie Chart
plt.subplot(1, 2, 2)
wedges, texts, autotexts = plt.pie(value_counts.values,
                                   labels=[f'Skor {k}' for k in value_counts.index],
                                   autopct='%1.1f%%',
                                   colors=colors[2:5],
                                   startangle=90)
plt.title('Persentase Skor Kemudahan Beranda', fontsize=14, fontweight='bold')

# Perbesar font persentase
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.tight_layout()
plt.show()
```

```python
# Analisis tingkat kepuasan
satisfied = value_counts.get(4, 0) + value_counts.get(5, 0)
neutral = value_counts.get(3, 0)
dissatisfied = value_counts.get(1, 0) + value_counts.get(2, 0)

satisfaction_rate = (satisfied / len(df)) * 100
neutral_rate = (neutral / len(df)) * 100
dissatisfaction_rate = (dissatisfied / len(df)) * 100

print(" ANALISIS TINGKAT KEPUASAN BERANDA")
print("=" * 50)
print(f" SANGAT MUDAH (Skor 5): {value_counts.get(5, 0)} responden ({(value_counts.get(5, 0)/len(df)*100):.1f}%)")
print(f" MUDAH (Skor 4): {value_counts.get(4, 0)} responden ({(value_counts.get(4, 0)/len(df)*100):.1f}%)")
print(f"  CUKUP (Skor 3): {neutral} responden ({neutral_rate:.1f}%)")
print(f" SULIT (Skor 1-2): {dissatisfied} responden ({dissatisfaction_rate:.1f}%)")

print(f"\n TOTAL KEPUASAN:")
print(f" PUAS (Skor 4-5): {satisfied} responden ({satisfaction_rate:.1f}%)")
print(f"  NETRAL (Skor 3): {neutral} responden ({neutral_rate:.1f}%)")
```

```python
#Analisis 4 : Informasi Produk

# Visualisasi 1: Bar Chart
plt.figure(figsize=(12, 5))

# Warna untuk visualisasi - tema produk/segar
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']

plt.subplot(1, 2, 1)
bars = plt.bar(value_counts.index, value_counts.values,
               color=colors[2:5], edgecolor='black', alpha=0.8)
plt.title('Distribusi Skor Kelengkapan Informasi Produk', fontsize=14, fontweight='bold')
plt.xlabel('Skor Kelengkapan (1-5)', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
plt.xticks([3, 4, 5])

# Tambahkan nilai di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# Visualisasi 2: Pie Chart dengan emphasis
plt.subplot(1, 2, 2)
wedges, texts, autotexts = plt.pie(value_counts.values,
                                   labels=[f'Skor {k}' for k in value_counts.index],
                                   autopct='%1.1f%%',
                                   colors=colors[2:5],
                                   startangle=90,
                                   explode=[0.1 if score == 3 else 0 for score in value_counts.index])

plt.title('Persentase Skor Kelengkapan Informasi Produk', fontsize=14, fontweight='bold')

# Perbesar font persentase
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.tight_layout()
plt.show()

```
