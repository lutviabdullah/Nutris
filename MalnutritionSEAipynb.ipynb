```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulated nutrition data for South East Asian countries (2023)
data = {
    'Country': ['Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Myanmar',
                'Philippines', 'Singapore', 'Thailand', 'Vietnam', 'Timor-Leste'],
    'Calories_kcal': [1800, 1900, 1750, 2100, 1700, 1850, 2200, 2050, 1950, 1650],
    'Protein_g': [45, 50, 42, 60, 40, 48, 65, 55, 52, 38],
    'Zinc_mg': [6.0, 6.5, 5.8, 7.2, 5.5, 6.2, 7.5, 7.0, 6.8, 5.2],
    'VitaminA_ug': [350, 400, 300, 450, 320, 380, 500, 420, 410, 290]
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. Basic Statistics
print("Average Nutrition Intake:")
print(df[['Calories_kcal', 'Protein_g', 'Zinc_mg', 'VitaminA_ug']].mean().round(1))

# 2. Deficiency Analysis
# WHO Recommended Minimums: Calories=1800, Protein=50g, Zinc=6mg, VitaminA=400ug
deficiency = pd.DataFrame({
    'Calories_def': (df['Calories_kcal'] < 1800).mean() * 100,
    'Protein_def': (df['Protein_g'] < 50).mean() * 100,
    'Zinc_def': (df['Zinc_mg'] < 6).mean() * 100,
    'VitaminA_def': (df['VitaminA_ug'] < 400).mean() * 100
}, index=['Deficiency Rate (%)'])

print("\nDeficiency Rates:")
print(deficiency)

# 3. Visualization
plt.figure(figsize=(15, 10))

# Calories and Protein Intake
plt.subplot(2, 2, 1)
sns.barplot(x='Country', y='Calories_kcal', data=df, palette='viridis')
plt.axhline(y=1800, color='r', linestyle='--', label='Minimum Requirement')
plt.title('Daily Calorie Intake (kcal)')
plt.xticks(rotation=45)
plt.legend()

plt.subplot(2, 2, 2)
sns.barplot(x='Country', y='Protein_g', data=df, palette='viridis')
plt.axhline(y=50, color='r', linestyle='--', label='Minimum Requirement')
plt.title('Protein Intake (g)')
plt.xticks(rotation=45)
plt.legend()

# Micronutrients
plt.subplot(2, 2, 3)
sns.barplot(x='Country', y='Zinc_mg', data=df, palette='viridis')
plt.axhline(y=6, color='r', linestyle='--', label='Minimum Requirement')
plt.title('Zinc Intake (mg)')
plt.xticks(rotation=45)
plt.legend()

plt.subplot(2, 2, 4)
sns.barplot(x='Country', y='VitaminA_ug', data=df, palette='viridis')
plt.axhline(y=400, color='r', linestyle='--', label='Minimum Requirement')
plt.title('Vitamin A Intake (Âµg)')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# 4. Correlation Matrix (Optional)
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Nutrient Intake Correlation')
plt.show()
```
